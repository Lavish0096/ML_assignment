```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('AWCustomers.csv')
print(df.head())
selected_columns = [
    'BirthDate', 'Gender', 'YearlyIncome', 'Occupation', 'MaritalStatus',
    'NumberChildrenAtHome', 'PostalCode',
    'NumberCarsOwned', 'TotalChildren','MiddleName'
]

df_selected = df[selected_columns]
print(df_selected.head())
df_selected['MiddleName'].fillna(df_selected['MiddleName'].mode()[0], inplace=True)
df_selected['YearlyIncome'] = df_selected['YearlyIncome'] / df_selected['YearlyIncome'].max()
df_selected['NumberCarsOwned'] = df_selected['NumberCarsOwned'] / df_selected['NumberCarsOwned'].max()

print(df_selected.head())
df_selected['BirthDate'] = pd.to_datetime(df_selected['BirthDate'])
df_selected['Age'] = (pd.Timestamp.now() - df_selected['BirthDate']).dt.days / 365.25
df_selected['Age'] = pd.cut(df_selected['Age'], bins=[0, 25, 40, 65, 100], labels=['Young', 'Adult', 'Middle Age', 'Senior'])
data = {
    'BirthDate': ['1990-05-14', '1985-07-22'],
    'Gender': ['M', 'F'],
    'YearlyIncome': [50000, 60000],
    'Occupation': ['Engineer', 'Doctor'],
    'MaritalStatus': ['Single', 'Married'],
    'NumberChildrenAtHome': [0, 2],
    'PostalCode': ['12345', '67890'],
    'NumberCarsOwned': [1, 2],
    'TotalChildren': [0, 2],
    'MiddleName': ['John', 'Doe']
}

df_selected = pd.DataFrame(data)
df_selected['BirthDate'] = pd.to_datetime(df_selected['BirthDate'])
df_selected['Age'] = (pd.Timestamp.now() - df_selected['BirthDate']).dt.days / 365.25
df_selected['Age'] = pd.cut(df_selected['Age'], bins=[0, 25, 40, 65, 100], labels=['Young', 'Adult', 'Middle Age', 'Senior'])
label_encoders = {}
for column in ['Gender', 'Occupation', 'MaritalStatus', 'PostalCode', 'MiddleName', 'Age']:
    le = LabelEncoder()
    df_selected[column] = le.fit_transform(df_selected[column])
    label_encoders[column] = le

# Convert 'BirthDate' to a numerical representation (days since a reference date)
# You can choose a reference date that makes sense for your data
reference_date = pd.Timestamp('1900-01-01')
df_selected['BirthDate'] = (df_selected['BirthDate'] - reference_date).dt.days

obj1 = df_selected.iloc[0]
obj2 = df_selected.iloc[1]
def simple_matching(obj1, obj2):
    same = (obj1 == obj2).sum()
    total = len(obj1)
    return same / total

smc = simple_matching(obj1, obj2)
print(f"Simple Matching Coefficient: {smc:.2f}")
def jaccard_similarity(obj1, obj2):
    set1 = set(obj1)
    set2 = set(obj2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

jaccard_sim = jaccard_similarity(obj1, obj2)
print(f"Jaccard Similarity: {jaccard_sim:.2f}")
```
